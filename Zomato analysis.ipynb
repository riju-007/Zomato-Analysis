```python
def price_bucket(x):
    if x<=10: 
        return "0-10"
    elif x >10 and x<=20:
        return "10-20"
    elif x >20 and x<=30:
        return "20-30"
    elif x >30 and x<=40:
        return "30-40"
    elif x >40 and x<=50:
        return "40-50"
    else:
        return "above 50"
```

```python
main_1['Price_bucket']=main_1['USD_Average_Cost_for_two'].apply(price_bucket)
```

```python
def rating_bucket(x):
    if x<=2:
        return "1-2"
    elif x>2 and x<=3:
        return "2-3"
    elif x>3 and x<=4:
        return "3-4"
    else:
        return "4-5"
```

```python
main_1['Rating_bucket']=main_1['Rating'].apply(rating_bucket)
main_1.head()
```

```python
main_1['Year'] = main_1['datekey'].dt.year
main_1['Month_no.'] = main_1['datekey'].dt.month
main_1['Month_name'] = main_1['datekey'].dt.strftime('%B')
main_1['Year_Month'] = main_1['datekey'].dt.strftime('%Y-%b')
main_1.head()
```

```python
def qtr(x):
    if x in [1,2,3]:
        return 'Q1'
    elif x in [4,5,6]:
        return 'Q2'
    elif x in [7,8,9]:
        return 'Q3'
    else:
        return 'Q4'
```

```python
main_1['Qtr']=main_1['Month_no.'].apply(qtr)
main_1.head()
```

```python
main_1['Weekday_no.']=main_1['datekey'].dt.weekday
main_1['Weekday']=main_1['datekey'].dt.strftime('%A')
main_1.head()
```

```python
def day_end(x):
    if x in [5,6]:
        return 'weekend'
    else:
        return 'weekday'
```

```python
main_1['Weekday/Weekend']=main_1['Weekday_no.'].apply(day_end)
main_1.head()
```

```python
def fcl_qtr(x):
    if x in [4,5,6]:
        return 'FQ1'
    elif x in [7,8,9]:
        return 'FQ2'
    elif x in [10,11,12]:
        return 'FQ3'
    else:
        return 'FQ4'
```

```python
main_1['Financial_Qtr']=main_1['Month_no.'].apply(fcl_qtr)
main_1.head()
```

```python
def fcl_mnth(x):
    if x==4:
        return 'FM1'
    elif x==5:
        return 'FM2'
    elif x==6:
        return 'FM3'
    elif x==7:
        return 'FM4'
    elif x==8:
        return 'FM5'
    elif x==9:
        return 'FM6'
    elif x==10:
        return 'FM7'
    elif x==11:
        return 'FM8'
    elif x==12:
        return 'FM9'
    elif x==1:
        return 'FM10'
    elif x==2:
        return 'FM11'
    else:
        return 'FM12'
```

```python
main_1['Financial_Month']=main_1['Month_name'].apply(fcl_mnth)
main_1.head()
```

```python
#Count of Resturants based on Ratings Bucket
rating_bucket = main_1.groupby('Rating_bucket')['RestaurantID'].count()
rating_bucket
```

```python
avg_rating = pd.DataFrame(rating_bucket)
avg_rating
```

```python
# Plotting a pie chart with custom colors and count labels
fig, ax = plt.subplots(figsize=(4, 4))

# Extract data for plotting
sizes = avg_rating['RestaurantID'].values
labels = avg_rating.index

# Define a custom color palette
custom_colors = ['blue', 'coral', 'red', 'green']

# Plot pie chart with custom colors
patches, texts, autotexts = ax.pie(sizes, labels=labels, autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p/100*sum(sizes)),
                                   startangle=90, colors=custom_colors, textprops=dict(color="w"))

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

# Customize text properties
for text, autotext in zip(texts, autotexts):
    text.set_size(12)
    autotext.set_size(10)

plt.title('Restaurant Count by Rating Bucket')
plt.show()
```

```python
# Plotting a pie chart with custom colors and adjusted count labels
fig, ax = plt.subplots(figsize=(4, 4))

# Extract data for plotting
sizes = price_bucket['RestaurantID'].values
labels = price_bucket.index

# Define a custom color palette
custom_colors = ['red', 'coral', 'black', 'green', 'violet', 'orange']

# Set a threshold for displaying data labels (e.g., 5%)
threshold = 5

# Plot pie chart with custom colors
patches, texts, autotexts = ax.pie(sizes, labels=labels,
                                   autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p/100*sum(sizes)) if p > threshold else '',
                                   startangle=90, colors=custom_colors, textprops=dict(color="w"))

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

# Customize text properties
for text, autotext in zip(texts, autotexts):
    text.set_size(12)
    autotext.set_size(10)

plt.title('Restaurant Count by Price Bucket')
plt.show()
```
